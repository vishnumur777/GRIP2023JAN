# **Renaming Column Names of the dataset**

```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)
```

```python
df.head(2)
```

# **Filling NaN values to 0 for Wound and Kill columns**

```python
df['Wound']=df['Wound'].fillna(0)
df['kill']=df['kill'].fillna(0)
```

```python
df['Casualities']=df['kill']+df['Wound']
```

# **Bar Chart representing Attack in years**

```python
year = df['Year'].unique()
years_count = df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = year,
           y = years_count,
           palette = "flare")
plt.xticks(rotation = 50)
plt.xlabel('Attacking Year',size=20,fontname="Eczar Semibold",fontweight="bold")
plt.ylabel('Number of Attacks Each Year',fontsize=20,fontweight="bold")
plt.title('Attacks In Years',fontname="Open Sans",size=40,fontweight="bold")
plt.show()
```

```python
import random
pd.crosstab(df.Year, df.Region).plot(kind='area',stacked=False,figsize=(20,10),color=["skyblue","green","red","orange","violet","magenta","yellow","grey","blue","lightgreen","aqua","pink"])
plt.title('Terrorist Activities By Region In Each Year',fontsize=25,fontweight="bold")
plt.ylabel('Number of Attacks',fontsize=20,fontweight="bold")
plt.xlabel("Year",fontsize=20,fontweight="bold")
plt.show()
```

```python
df1 = df[['Year','kill']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(20,10))
df1.plot(kind='bar',alpha=0.7,ax=ax4,color="red")
plt.xticks(rotation = 50)
plt.title("People Died Due To Attack",fontsize=25,weight="bold")
plt.ylabel("Number of killed peope",fontsize=20)
plt.xlabel('Year',fontsize=20)
top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

```python
df['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color='darkorange')
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Top 10 most effected city",fontsize=20,weight="bold")
plt.show
```

```python
df['Attacktype'].value_counts().plot(kind='bar',figsize=(20,10),color='#005aff')
plt.xticks(rotation = 50)
plt.xlabel("Attacktype",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Name of attacktype",fontsize=30,weight="bold")
plt.show()
```

```python
df[['Attacktype','kill']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,10),color='#a1b2c3')
plt.xticks(rotation=50)
plt.title("Number of killed ",fontsize=20,weight="bold")
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=15)
plt.show()
```

```python
df[['Attacktype','Wound']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,10),color='#d4e500')
plt.xticks(rotation=50)
plt.title("Number of wounded  ",fontsize=20,weight="bold")
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=15)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.countplot(df["Targettype"],order=df['Targettype'].value_counts().index,palette="icefire",edgecolor=sns.color_palette("Reds"));
plt.xticks(rotation=90)
plt.xlabel("Attacktype",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.title("Attack per year",fontsize=20,weight="bold")
plt.show()
```

```python
df['Group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='#2B76E5',figsize=(20,10))
plt.title("Top 10 terrorist group attack",fontsize=20,weight="bold")
plt.xlabel("terrorist group name",fontsize=15)
plt.ylabel("Attack number",fontsize=15)
plt.show()
```

```python
df[['Group','kill']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('kill',ascending=False).head(10).plot(kind='bar',color='#E987B5',figsize=(20,10))
plt.title("Top 10 terrorist group attack",fontsize=20,weight="bold")
plt.xlabel("terrorist group name",fontsize=15)
plt.ylabel("No of killed people",fontsize=15)
plt.show()
```
